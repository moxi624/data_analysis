```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 导入sklearn相关类库
# sklearn 不是做神经网络的，是做一些传统的机器学习算法
# 比如 逻辑回归，决策树，随机森林
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import KFold
```

```python
# 导入数据集
filename = "data.csv"
raw = pd.read_csv(filename)
print(raw.shape)
raw.head()
```

```python
# 500 for test
# 拿到数据后，需要把数据切分成两部分
# 下面有大约3w个数据，我们 把大部分划分成 train数据，少部分划分成test数据
# shot_made_flag 代表科比投篮有没有进， 其中如果有标签的，代表是训练集，没有数据的就是预测集了。
kobe= raw[pd.notnull(raw['shot_made_flag'])]
print(kobe.shape)
```

```python
# 画图操作
alpha = 0.02  # aloha表示透明程度

plt.figure(figsize=(10, 10))

# loc_x  和 loc_y
# 下面是查看一些指标  loc_x, loc_y, lon, lat分别代表着什么含义
# scatter是表示散点图
plt.subplot(121)  # 121  表示布局   12：画的东西在 1行2列   1：
plt.scatter(kobe.loc_x, kobe.loc_y, color="R", alpha=alpha)
plt.title('locx and loc_y')

# lat 和 lon  表示经度和维度
plt.subplot(122)
plt.scatter(kobe.lon, kobe.lat, color='B', alpha=alpha)
plt.title('lat and lon')

# 下面的图，表示科比投篮位置
# 从下面的图可以知道，经纬度  和 xy轴是一个线性相关的特征值，所以我们只需要保留一个即可。
```

```python
# 下面是进行一些特征转换
# 通过 loc_x 和 loc_y  转换成极坐标系  得到 角度 和 距离
# 在我们拿到一份数据的时候，首先需要想到的是，特征的提取
# 1、首先计算欧式距离
raw['dist'] = np.sqrt(raw['loc_x']**2 + raw['loc_y']**2)

loc_x_zero = raw['loc_x'] == 0
# print(loc_x_zero)
raw['angle'] = np.array([0]*len(raw))
raw['angle'][~loc_x_zero] = np.arctan(raw['loc_y'][~loc_x_zero] / raw['loc_x'][~loc_x_zero])
raw['angle'][loc_x_zero] = np.pi / 2
# 最终得到了两个新的特征   dist 和 angle
```

```python
gs = kobe.groupby('shot_zone_area')
print(kobe['shot_zone_area'].value_counts())
print(len(gs))
```

```python
import matplotlib.cm as cm
plt.figure(figsize=(20,10))
def scatter_plot_by_category(feat):
    alpha = 0.1
    gs = kobe.groupby(feat)
    cs = cm.rainbow(np.linspace(0, 1, len(gs)))
    for g, c in zip(gs, cs):
        plt.scatter(g[1].loc_x, g[1].loc_y, color=c, alpha=alpha)

# shot_zone_area
plt.subplot(131)
scatter_plot_by_category('shot_zone_area')
plt.title('shot_zone_area')

# shot_zone_basic
plt.subplot(132)
scatter_plot_by_category('shot_zone_basic')
plt.title('shot_zone_basic')

# shot_zone_range
plt.subplot(133)
scatter_plot_by_category('shot_zone_range')
plt.title('shot_zone_range')
```

```python
# 增加需要删除的字段
drops = ['shot_id', 'team_id', 'team_name', 'shot_zone_area', 'shot_zone_range', 'shot_zone_basic', 'matchup', 'lon', 'lat', 'seconds_remaining', \
        'minutes_remaining', 'shot_distance', 'loc_x', 'loc_y', 'game_event_id', 'game_id', 'game_date']

for drop in drops:
    raw = raw.drop(drop, 1)  # 1：代表删除一列
```

```python
categorical_vars = ['action_type', 'combined_shot_type', 'shot_type', 'opponent', 'period', 'season']
for var in categorical_vars:
    print(var)
    raw = pd.concat([raw, pd.get_dummies(raw[var], prefix=var)], 1)  # concat是拼接的操作
    raw = raw.drop(var, 1) # 删除var
```

```python
train_kobe = raw[pd.notnull(raw['shot_made_flag'])]
train_label = train_kobe['shot_made_flag']
train_kobe = train_kobe.drop('shot_made_flag', 1)
# test_kobe = raw[pd.isnull(raw['shot_made_flag'])]
# test_kobe = test_kobe.drop('shot_made_flag', 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, log_loss
import time
```

```python
import numpy as np
range_m = np.logspace(0, 2, num=5).astype(int)  # 转换成int类型 
range_m
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import KFold

print("Finding best n_estimators for RandomForestClassifier")
min_score = 100000
best_n = 0
scores_n = []
range_n = np.logspace(0, 2, num=3).astype(int)
for n in range_n:
    print("the number of trees:{0}".format(n))
    t1 = time.time()
    
    rfc_score = 0
    
    # 选择随机森林的分类器，传入n，表示多少个树
    rfc = RandomForestClassifier(n_estimators=n)
    
    for train_k, test_k in KFold(len(train_kobe), n_folds=10, shuffle=True):
        
        # 使用模型 fit 一下数据， 需要传入训练数据和标签数据
        rfc.fit(train_kobe.iloc[train_k], train_label.iloc[train_k])
        
        # 预测新的结果
        pred = rfc.predict(train_kobe.iloc[test_k])
        
        # 对模型进行评测（召回率）
        rfc_score += log_loss(train_label.iloc[test_k], pred) / 10
    
    scores_n.append(rfc_score)
    
    # 找出最合适的参数
    if rfc_score < min_score:
        min_score = rfc_score
        best_n = n
        
    t2 = time.time()
    print("Done procession {0} trees ({1:.3f}sec)".format(n, t2 - t1))

print(best_n, min_score)

scores_n
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(rang_n, scores_n)
plt.ylabel('score')
plt.xlabel('number of trees')

plt.subplot(122)
plt.plot(range_m, scores_m)
plt.ylabel('score')
plt.xlabel('max depth')
```
