```python
import torch
import numpy as np
import pandas as pd
import torch.nn as nn
import torch.nn.functional as F

import seaborn as sns
```

```python
pf = pd.read_csv('data/FEdata.csv')
```

```python
pf = pf.drop(['Over18','StandardHours'],axis=1)
```

```python
pf.head()
```

```python
DepartmentDf = pd.get_dummies( pf['Department'] , prefix='Department' )
DepartmentDf.head()
```

```python
pf = pd.concat([pf,DepartmentDf], axis=1)
```

```python
pf.drop('Department',axis=1,inplace=True)
```

```python
pf.head()
```

```python
EducationFieldDf = pd.get_dummies( pf['EducationField'] , prefix='EducationField' )
pf = pd.concat([pf,EducationFieldDf], axis=1)
pf.drop('EducationField',axis=1,inplace=True)
pf.head()
```

```python
JobRoleDf = pd.get_dummies( pf['JobRole'] , prefix='JobRole' )
pf = pd.concat([pf,JobRoleDf], axis=1)
pf.drop('JobRole',axis=1,inplace=True)
pf.head()
```

```python
pf.info()
```

```python
pf.head()
```

```python
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(pf, pf['Attrition'], test_size=0.2)
```

```python
y1_train = torch.from_numpy(np.array(y1_train))
y1_test = torch.from_numpy(np.array(y1_test))
```

```python
y1_train.size
```

```python
y1_train = y1_train.reshape(880,1)
y1_test = y1_test.reshape(220,1)
```

```python
y1_train = y1_train.float()
y1_test = y1_test.float()
```

```python
x1_train = x1_train.drop(['Attrition'],axis=1)
x1_test = x1_test.drop(['Attrition'],axis=1)
```

```python
x1_train_np = np.array(x1_train)
x1_test_np = np.array(x1_test)
```

```python
x1_train_np = x1_train_np/x1_train_np.max(axis=0)
x1_test_np = x1_test_np/x1_test_np.max(axis=0)
```

```python
x1_train = torch.from_numpy(x1_train_np)
x1_test = torch.from_numpy(x1_test_np)
```

```python
x1_train = x1_train.float()
x1_test = x1_test.float()
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.layer1 = nn.Sequential(
        nn.Linear(43,256),
        #nn.Dropout(p=0.2),
        nn.Linear(256,128),
        #nn.Dropout(p=0.2),
        nn.Linear(128,32),
        #nn.Dropout(p=0.2),
        nn.Linear(32,1))
        
    def forward(self, x):
        x = self.layer1(x)
        return x
    
net = Net()

print(net)
```

```python
optimizer = torch.optim.Adam(net.parameters(), lr=0.0001)
loss_func = torch.nn.MSELoss() 

for i in range(1000):
    out = net(x1_train)
    loss = loss_func(out,y1_train)
    if i%50==0:
        testout = net(x1_test)
        testloss = loss_func(testout,y1_test)
        print(testloss)
        print(i)
    optimizer.zero_grad()   # 清空上一步的残余更新参数值
    loss.backward()         # 误差反向传播, 计算参数更新值
    optimizer.step()        # 将参数更新值施加到 net 的 parameters 上
```

```python
pf_pre = pd.read_csv('data/FEtest.csv')
pf_pre = pf_pre.drop(['Over18','StandardHours'],axis=1)
EducationFieldDf = pd.get_dummies( pf_pre['EducationField'] , prefix='EducationField' )
pf_pre = pd.concat([pf_pre,EducationFieldDf], axis=1)
pf_pre.drop('EducationField',axis=1,inplace=True)

DepartmentDf = pd.get_dummies( pf_pre['Department'] , prefix='Department' )
pf_pre = pd.concat([pf_pre,DepartmentDf], axis=1)
pf_pre.drop('Department',axis=1,inplace=True)

JobRoleDf = pd.get_dummies( pf_pre['JobRole'] , prefix='JobRole' )
pf_pre = pd.concat([pf_pre,JobRoleDf], axis=1)
pf_pre.drop('JobRole',axis=1,inplace=True)

pfnp_pre = np.array(pf_pre)
pfnp_pre = pfnp_pre/pfnp_pre.max(axis=0)
x_pre = torch.from_numpy(pfnp_pre)
x_pre.size()
```

```python
x_pre = x_pre.float()
pre = net(x_pre)
```

```python
pre = pre.detach().numpy()
```

```python
for x in range(len(pre)):
    if pre[x]>0.5:
        pre[x]=1
    else:
        pre[x]=0
pre
```
